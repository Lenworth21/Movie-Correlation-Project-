```python
#import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)# Adjusts the configuration of the plots we will create

#read in the data
df = pd.read_csv(r'C:\Users\Lenworth\Documents\Life After College\movies.csv')
```

```python
#Looking at the data
df.head()
```

```python
#Checking to see if there is any missing data using for loop
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#Data types for the columns
df.dtypes
```

```python
#Changed data type of column from a float to an integer 
df['gross'] = df['gross'].astype("Int64")
df['budget'] = df['budget'].astype("Int64")
```

```python
df.head()
```

```python
#Create correct Year Column
df['yearcorrect'] = df['released'].astype(str).str[:4]
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Remove any duplicates that are present
df.drop_duplicates()
df.head()
```

```python
#budget high correlation
#company high correlation
```

```python
plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
matplotlib.pyplot.xlabel('Gross')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot budget vs gross using seaborn 
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"Color":"blue"})
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
#looks at Company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
        
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs= correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr= sorted_pairs[(sorted_pairs)> 0.5 ]
high_corr
```
